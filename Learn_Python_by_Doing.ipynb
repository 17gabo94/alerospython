# üß± 1Ô∏è‚É£ Object-Oriented Programming (OOP)


So far, you‚Äôve learned how to work with variables, loops, and functions.
Now it‚Äôs time to explore one of the most powerful ideas in programming: **Object-Oriented Programming** (OOP).

OOP helps us organize our code by grouping data and actions into something called **classes**.


üèóÔ∏è What is a Class?

A class is like a blueprint for creating objects.
An object is something created from that class ‚Äî like a car built from a car design.

```python
# Example: A simple class
class Dog:
    def __init__(self, name, age):
        # __init__ is the constructor ‚Äî it runs automatically when we create a Dog
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says: Woof! üê∂")

# Create two Dog objects
buddy = Dog("Buddy", 3)
luna = Dog("Luna", 5)

buddy.bark()
luna.bark()
```

üß± Understanding Constructors and Attributes
	‚Ä¢	Constructor (__init__): runs automatically when we create a new object.
	‚Ä¢	Attributes (self.name): store information about the object.
	‚Ä¢	Methods: functions that belong to a class.

üß¨ Inheritance

Inheritance allows one class to use the features of another class.
For example, a Cat class can inherit from an Animal class.

```python
# Base class
class Animal:
    def __init__(self, species):
        self.species = species

    def speak(self):
        print(f"The {self.species} makes a sound.")

# Derived class
class Cat(Animal):
    def __init__(self, name, age):
        super().__init__("Cat")  # Call the base class constructor
        self.name = name
        self.age = age

    def speak(self):
        print(f"{self.name} says: Meow! üê±")

# Using inheritance
milo = Cat("Milo", 2)
milo.speak()
milo2 = Animal("Milo2")
milo2.speak()
```

‚öôÔ∏è Properties (@property)

Properties allow you to control how values are read or modified in a safe way.
They help you protect data while keeping a clean syntax.

```python
class Person:
    def __init__(self, name, age):
        self._name = name
        self._age = age  # note the underscore ‚Äî it's a convention for "private" attributes

    @property
    def age(self):
        return self._age

    @age.setter
    def age(self, new_age):
        if new_age >= 0:
            self._age = new_age
        else:
            print("Age cannot be negative!")

person = Person("Alice", 25)
print(person.age)

person.age = 30  # valid
print(person.age)

person.age = -5  # invalid
```
