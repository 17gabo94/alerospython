
## 2Ô∏è‚É£ Variables and Data Types

A **variable** is like a box where you can store information.

You can store numbers, text, or True/False values.  
Each type of data is called a **data type**.


```python

# Examples of variables
age = 25
name = "Alice"
temperature = 18.5
is_sunny = True

print("Name:", name)
print("Age:", age)
print("Temperature:", temperature)
print("Is it sunny?", is_sunny)
```


## 3Ô∏è‚É£ Operators

Operators help us perform calculations or comparisons.

### Arithmetic Operators:
`+`, `-`, `*`, `/`, `//`, `%`, `**`

### Comparison Operators:
`==`, `!=`, `>`, `<`, `>=`, `<=`


```python

# Example: simple math and comparisons
apples = 5
bananas = 3

print("Total fruits:", apples + bananas)
print("Do we have more apples than bananas?", apples > bananas)
```


## 4Ô∏è‚É£ Conditional Statements (`if`, `elif`, `else`)

Sometimes we want our program to make decisions.

We can use `if` to run code **only if** a condition is true.


```python

temperature = 22

if temperature > 25:
    print("It's a hot day ‚òÄÔ∏è")
elif temperature > 15:
    print("It's a nice day üå§Ô∏è")
else:
    print("It's a bit cold üß£")
```


## 5Ô∏è‚É£ Loops (`for` and `while`)

A **loop** repeats a block of code several times.

### `for` loop ‚Äî used when we know how many times to repeat.
### `while` loop ‚Äî repeats while a condition is true.


```python

# Example 1: using a for loop
for number in range(1, 6):
    print("Counting:", number)

# Example 2: using a while loop
counter = 1
while counter <= 5:
    print("While counting:", counter)
    counter += 1
```


## 6Ô∏è‚É£ Lists and Dictionaries

A **list** stores multiple values in order.  
A **dictionary** stores key-value pairs, like a mini database.


```python

# List example
fruits = ["apple", "banana", "cherry"]
print("My fruits:", fruits)

# Dictionary example
person = {"name": "Alice", "age": 25, "city": "Dublin"}
print("Person info:", person)
```


## 7Ô∏è‚É£ Functions

A **function** is a reusable piece of code that performs a specific task.  
It helps us avoid repeating code.


```python

def greet_user(user_name):
    print(f"Hello, {user_name}! Welcome to Python learning üòä")

greet_user("Alice")
```


## 8Ô∏è‚É£ Error Handling

Sometimes our code might fail ‚Äî maybe we divide by zero or use a wrong variable.  
We can use `try` and `except` to handle those errors gracefully.


```python

try:
    number = int(input("Enter a number: "))
    result = 10 / number
    print("Result:", result)
except ZeroDivisionError:
    print("Oops! You can't divide by zero.")
except ValueError:
    print("Please enter a valid number.")
```
